# TF-IDF

```python
#term frequency = frequency of the word in this doc/all terms in this doc
keys0, keys1, keys2 = {},{},{}
for j in range(3):
    for i in globals()["vol"+str(j)]:
        if i not in globals()["keys"+str(j)]:
            globals()["keys"+str(j)][i] = 1
        else:
            globals()["keys"+str(j)][i] += 1
    
```

```python
keys2
```

```python
#tf
tfs0, tfs1, tfs2 = {},{},{}
for j in range(3):
    for i in globals()["keys"+str(j)]:
        globals()["tfs"+str(j)][i] = 1/len(globals()["vol"+str(j)])
```

```python
tfs0
```

```python
#idf
idf={}
from collections import Counter
X, Y, Z = Counter(keys0),Counter(keys1), Counter(keys2)
idf = dict(X+Y+Z)
for i in idf:
    if(i in keys0 and i in keys1 and i in keys2 ):
        idf[i] = math.log(3/3)
    elif((i in keys0 and i not in keys1 and i not in keys2) or (i not in keys0 and i in keys1 and i not in keys2) or (i not in keys0 and i not in keys1 and i in keys2) ):
        idf[i] = math.log(3/2)
    else:
        idf[i] - math.log(3/1)
```

```python
tf_idf0,tf_idf1,tf_idf2 = keys0,keys1,keys2
for i in range(3): 
    for j in globals()["keys"+str(i)]:
        (globals()["tf_idf"+str(i)])[j] = (globals()["tfs"+str(i)])[j]*(idf[j])
```
